```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv(r'/content/banking_data.csv')
```

```python
pd.set_option('display.max_rows',None)
pd.set_option('display.max_columns',None)
print("BEFORE:\n")
print(df.isnull().sum())
# print(df['marital_status'].value_counts())
# print(df['marital'].value_counts())
```

```python
df.drop(columns=['marital'],inplace=True)
df['marital_status']=df['marital_status'].fillna(df['marital_status'].mode()[0])
df['education']=df['education'].fillna(df['education'].mode()[0])
print("AFTER:\n")
print(df.isnull().sum())
```


#What is the distribution of age among the clients?

```python
df['age'].hist(bins=20)
plt.hist(df['age'],color='skyblue',edgecolor='black')
plt.xlabel('Age',color='blue',size=12)
plt.ylabel('Number of Clients',color='blue',size=12)
plt.title('Distribution by Age',color='red',size=15)
plt.show()
```

#Are there any correlations between different attributes and the likelihood of subscribing to a term deposit?

```python
df['term_deposit']=df['term_deposit'].replace({'yes':1,'no':0})
df['default']=df['default'].replace({'yes':1,'no':0})
df['housing']=df['housing'].replace({'yes':1,'no':0})
df['loan']=df['loan'].replace({'yes':1,'no':0})
# df['month']=df['month'].replace({'jan':1,'feb':2,'mar':3,'apr':4,'may':5,'jun':6,'jul':7,'aug':8,'sep':9,'oct':10,'nov':11,'dec':12})
numeric_df=df.select_dtypes(include=['int64','float64'])
corr_matrix=numeric_df.corr()
plt.figure(figsize=(10,8))
sns.heatmap(corr_matrix,annot=True,cmap='PuBuGn',fmt=".2f")
plt.title('Correlation Matrix denoting Correlations between Different Attributes and the Likelihood of Subscribing to a Term Deposit',color='red',size=15)
plt.show()
```
